```python
import pandas as pd
import numpy as np
```

```python
#Função que cria series

def gerador_serie_historica(valor_inicial,volatilidade,periodos,dia_inicial,frequencia = "M"):
   
   vetor = [valor_inicial]
   for i in range(periodos -1):
        preco = vetor[i] * (1 + np.random.normal(0,volatilidade))

        vetor.append(preco)

   serie = pd.Series(vetor,index = pd.date_range(dia_inicial,periods = periodos,
                                                  freq = frequencia ))
   
   return serie
```

```python
serie_lucro = gerador_serie_historica(2,0.5,5,"2022-01-01")
serie_cotacao = gerador_serie_historica(20,0.5,5,"2022-01-01")
df = pd.DataFrame({"Cotação" : serie_cotacao,
                "Lucro por ação" : serie_lucro})

df["PL"] = df["Cotação"]/df["Lucro por ação"]

df
```

```python
#df_base

dicionario = {
       "nomes": ["Weg", "Petrobras", "Vale", "Petrobras", "Lojas Renner"],
       "tickers": ["WEGE3", "PETR3", "VALE3", "PETR4", "LREN3"],  
       "cotacoes": [20, 30, 40, 12, 35],
       "preco_sobre_lucro": [25, 6, 12, 7, 25],  
       "volume": [5000, 1000, 4000, 7000, 1200]
}

df_info_empresas = pd.DataFrame(dicionario)

df_info_empresas = df_info_empresas.set_index("tickers")

df_info_empresas.columns = ['nomes', 'precos', 'preco_sobre_lucro', 'volume']

df_info_empresas['lpa'] = df_info_empresas['precos'] / df_info_empresas['preco_sobre_lucro']

dolar = 5.25

df_info_empresas['precos_dolar'] = df_info_empresas['precos'] / dolar

df_info_empresas
```

```python
# 53 Ordene o índice do dataframe

df_info_empresas = df_info_empresas.sort_index()

df_info_empresas
```

```python
# 54 Ordene a tabela de forma decrescente a partir do indicador PL

df_info_empresas = df_info_empresas.sort_values(by="preco_sobre_lucro", ascending= False)


df_info_empresas
```

```python
#55 Oredene a tabela de forma crescente a partir das colunas PL e volume, com o volume desampatado

df_info_empresas = df_info_empresas.sort_values(by = ["preco_sobre_lucro","volume"])

df_info_empresas
```

```python
# 56 Crie uma coluna com o ranking de LPA das empresas, utilizadno o método que preferir(Max,Min,Dense...)

df_info_empresas["Ranking-LPA"] = df_info_empresas["lpa"].rank()

df_info_empresas = df_info_empresas.sort_values(by ="Ranking-LPA")

df_info_empresas
```
